```python
from collections import Counter

N, K = map(int, input().split())
a = input().split()
b = []
c = a.copy()


plug = [0] * N
sum = 0

if N >= K:
    print('0')


for i in c:
    # print(i)

    if 0 in plug:
        print('0 있음')        
        plug[plug.index(0)] = a[0]
        del a[0]
        print(plug)
        print(a)

    elif 0 not in plug and i in plug:
        print('있음')
        a.pop(0)
        print(plug)
        print(a)

    else:
        print('없음')
        cnt = Counter(a)
        for j in plug:
            b.append(cnt[j])
        plug[b.index(min(b))] = a.pop(0)
        print(plug)
        print(a)
        b.clear()
        sum += 1

    if len(a) == 0:
        break
print(sum)

```
